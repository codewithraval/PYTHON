```python
# display a frequency distribution for personal Loan

import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
personal_loan_counts = df['Personal Loan'].value_counts()
personal_loan_percentages = df['Personal Loan'].value_counts(normalize=True) * 100

personal_loan_df = pd.DataFrame({
    'Personal Loan Status': personal_loan_counts.index.map({0: 'No Loan', 1: 'Took Loan'}),
    'Count': personal_loan_counts.values,
    'Percentage': personal_loan_percentages.values
})

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(data=personal_loan_df, x='Personal Loan Status', y='Percentage', palette='Set2')

# Annotate both count and percentage
for index, row in personal_loan_df.iterrows():
    plt.text(x=index, 
             y=row['Percentage'] + 2, 
             s=f"{row['Count']} ({row['Percentage']:.2f}%)", 
             ha='center', 
             fontdict={'size': 12, 'weight': 'bold'})

# Titles and labels
plt.title('Personal Loan Distribution (Count and %)', fontdict={'fontsize': 16, 'weight': 'bold'})
plt.xlabel('Personal Loan Status', fontdict={'fontsize': 12, 'weight': 'bold'})
plt.ylabel('Percentage', fontdict={'fontsize': 12, 'weight': 'bold'})
plt.ylim(0, 120)  # Adjusted to leave space for text
plt.tight_layout()
plt.show()
```
