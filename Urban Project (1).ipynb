```python
## we replace the value with mode becouse null data is 50% above 

df['preciptype'] = df['preciptype'].fillna(df['preciptype'].mode()[0])
```

```python
## stations column have 93% data is null that's why we drop the column

df.drop(['stations'], axis=1, inplace=True)
```

```python
## Condition_Code column replace the value with mode becouse null data is 50% above 


df['Condition_Code'] = df['Condition_Code'].fillna(df['Condition_Code'].mode()[0])
```

```python
## replace the with o becouse no ather value in this column 

df['snowdepth'] = df['snowdepth'].fillna(0)
```

## Data   visulization

```python
## i create a boxplot 

# Set the style
sns.set(style="whitegrid")

# Select relevant columns for the box plot
cols_to_plot = ['tempmax', 'tempmin', 'temp', 'humidity', 'Heat_Index']

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[cols_to_plot])
plt.title('Box Plot of Weather and Health Metrics')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

sns.set(style="whitegrid")

city_counts = df['City'].value_counts()
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='City', data=df, order=city_counts.index, palette='Set2')

for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=10, color='black')

plt.title("Number of Records per City", fontsize=16, fontweight='bold')
plt.xlabel("City", fontsize=12)
plt.ylabel("Record Count", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='temp', y='Health_Risk_Score', hue='Season', palette='Set2')
plt.title('Health Risk Score vs. Temperature by Season')
plt.xlabel('Temperature (Â°F)')
plt.ylabel('Health Risk Score')
plt.legend(title='Season')
plt.tight_layout()
plt.show()
```

```python
## i create a pie chart of city wise temp

import matplotlib.pyplot as plt

city_avg_temp = df.groupby('City')['temp'].mean()

plt.figure(figsize=(8, 8))
plt.pie(city_avg_temp, labels=city_avg_temp.index, autopct='%1.1f%%', startangle=140)
plt.title('City-wise Average Temperature')
plt.axis('equal')  
plt.show()
```

```python
## i create bar chart of Customer Count by Weekend vs Weekday.

# Set style
sns.set(style="whitegrid")

weekend_counts = df['Is_Weekend'].value_counts()
total = len(df)
percentages = (weekend_counts / total * 100).round(1)
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Is_Weekend', data=df, palette='pastel', order=[False, True])
for p, count in zip(ax.patches, weekend_counts.sort_index()):
    percent = percentages.loc[p.get_x() == 0] if p.get_x() == 0 else percentages.loc[p.get_x() == 1]
    label = f'{count} ({percent:.1f}%)'
    ax.annotate(label, (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')
plt.title("Customer Count by Weekend vs Weekday", fontsize=16, fontweight='bold')
plt.xlabel("Is Weekend", fontsize=12)
plt.ylabel("Number of Records", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Weekday", "Weekend"])
plt.tight_layout()
plt.show()
```

```python
##  i create hbar of Average Temperature by Day of the Week.

# Sort data by temperature for better visual clarity
sorted_df = df[['Day_of_Week', 'temp']].groupby('Day_of_Week').mean().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(sorted_df.index, sorted_df['temp'], color='skyblue')

# Add labels to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f'{width:.1f}Â°F', va='center', fontsize=10)

# Labels and title
plt.xlabel('Average Temperature (Â°F)', fontsize=12)
plt.ylabel('Day of the Week', fontsize=12)
plt.title('Average Temperature by Day of the Week', fontsize=16, fontweight='bold')

# Tidy layout
plt.tight_layout()
plt.show()
```

```python
## i create a bar char of columns categorical data

categ_columns = []
for col in df.columns:
    if df[col].nunique() <= 5 and df[col].dtype in ['object', 'bool']:
        categ_columns.append(col)

print(f'Categorical columns with <=5 unique values: {categ_columns}')

# Plotting
plt.figure(figsize=[16, 12])
plt.suptitle('BAR CHART OF CATEGORICAL DATA', fontsize=18, fontweight='bold')
plt.subplots_adjust(top=0.92, hspace=0.5, wspace=0.4)

for i, col in enumerate(categ_columns[:6]):  # Show up to 6 columns
    ax = plt.subplot(2, 3, i+1)
    sns.countplot(data=df, x=col, hue=col, palette='viridis', legend=False)
    
    total = len(df[col].dropna())
    
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 5), 
                    textcoords='offset points')
        ax.annotate(f'{height/total:.1%}', 
                    (p.get_x() + p.get_width() / 2., 0), 
                    ha='center', va='center', 
                    xytext=(0, -25),
                    textcoords='offset points',
                    color='black',
                    fontsize=10)

    plt.title(f'{col.upper()}', fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
