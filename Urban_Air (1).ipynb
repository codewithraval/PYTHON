## Data Visulization

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Season', y='temp', palette='coolwarm')
plt.title('Temperature Distribution Across Seasons')
plt.xlabel('Season')
plt.ylabel('Temperature (Â°F)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='temp', y='Health_Risk_Score', hue='Season', palette='Set2')
plt.title('Health Risk Score vs. Temperature by Season')
plt.xlabel('Temperature (Â°F)')
plt.ylabel('Health Risk Score')
plt.legend(title='Season')
plt.tight_layout()
plt.show()
```

```python
x=df['datetime']
y=df['temp']
plt.scatter(x,y)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Temperature', fontsize=15)
plt.show()
```

```python
x=df['feelslikemin']
y=df['feelslikemax']
plt.scatter(x,y)
plt.xlabel('Feelslikemin', fontsize=15)
plt.ylabel('Feelslikemax', fontsize=15)
plt.show()
```

```python
## i create a pie chart of city wise temp

import matplotlib.pyplot as plt

city_avg_temp = df.groupby('City')['temp'].mean()

plt.figure(figsize=(8, 8))
plt.pie(city_avg_temp, labels=city_avg_temp.index, autopct='%1.1f%%', startangle=140)
plt.title('City-wise Average Temperature')
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['datetime'], df['temp'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('Temperature (Â°F)')
plt.title('Daily Temperature Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python

ax = sns.countplot(x= 'Is_Weekend', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count Of Customer By Weekend")
plt.show()
```

```python

ax = sns.countplot(x= 'City', data = df)

ax.bar_label(ax.containers[0])
plt.title("Number of Records per City")
plt.show()
```

```python
# Select numeric columns
numeric_df = df.select_dtypes(include='number')

# Correlation of 'temp' with others (as DataFrame for heatmap)
temp_corr_df = numeric_df.corr()[['temp']].sort_values(by='temp', ascending=False)

# Plot the heatmap
plt.figure(figsize=(6, 10))
sns.heatmap(temp_corr_df, annot=True, cmap='coolwarm')
plt.title('Correlation of temp with Other Numeric Features')
plt.show()
```
